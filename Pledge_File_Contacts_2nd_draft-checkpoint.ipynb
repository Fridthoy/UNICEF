# K-means clustering

```python
# Scaling the variables
X1 = scale(df['Installment Amount'])
X2 = scale(df['Total Paid Amount'])
X3 = scale(df['Number Of Paid Installments'])
X = (np.array([X1,X2,X3])).T
```

```python
# Building and running the model
clustering = KMeans(n_clusters=5, random_state=5) #the random state is to yield the same initialization every time you run the code
clustering.fit(X)
```

# Plotting

```python
color_theme = np.array(['darkgray','lightsalmon','powderblue', 'aqua', 'greenyellow'])
plt.scatter(x=df['Installment Amount'], y=df['Total Paid Amount'], c=color_theme[clustering.labels_], s=50)
plt.title('K-means Clustering')
plt.xlabel('Installment Amount')
plt.ylabel('Total Paid Amount')
```

# K-modes using numerical and cathegorical data!

```python
df_all = df2.copy() #making a copy of df without the abonrmality 
```

```python
df_all = df_all.drop(labels=['Contact: Contact ID',
                                   'Contact: Giver ID','Recurring Donation: ID', 'Installment Amount Currency', 'Total Paid Amount Currency',
                                   'Date Established','Date Closed','First Donation Date','Last Donation Date', 'Installment Period'] , axis = 1)

#take away all NaNs from df
df_all = df_all.fillna(0)  #replacing NaN by 0
df_all['Open Ended Status'] = df_all['Open Ended Status'].astype('str')
df_all['Channel'] = df_all['Channel'].astype('str')
df_all['Campaign'] = df_all['Campaign'].astype('str')
df_all['Pledge Type'] = df_all['Pledge Type'].astype('str')

df_all.head(10) 
```

```python
#for i, val in enumerate(arr): 
  #  print(i, val, arr[i])
```

```python
# Make the numerical data categorical such that it can be used in K-modes

#Installment Amount
def categorizes_installment(amount):    #cathegorizes amount to different clusters
    if amount<50: 
        return 'under 50'
    if amount<200: 
        return '50-199 kr'
    if amount<400: 
        return '200-599 kr'
    if amount<1000: 
        return '400-1000 kr'
    else: 
        return 'above 1000 kr'    

#Total Paid Amount. Mean is 6085. Max is 139400.
def categorizes_total(amount): 
    if amount<50: 
        return 'under 200'
    if amount<1000: 
        return '200-1000 kr'
    if amount<5000: 
        return '1000-5000 kr'
    if amount<10000: 
        return '5000-10000 kr'
    else: 
        return 'above 10000 kr' 

# Number Of Paid Installments. Mean is 32. Max is 208
def categorizes_nr(nr): 
    if nr<1: 
        return 'less than 1 paid'
    if nr<10: 
        return '1-9 paid'
    if nr<30: 
        return '10-29 paid'
    if nr<100: 
        return '30-99 kr'
    else: 
        return 'above 100 kr'  # 30 paiments in this category

    
df_fit = df_all.copy()
df_fit['Installment Amount'] = list(map(categorizes_installment,df_fit['Installment Amount']))
df_fit['Total Paid Amount'] = list(map(categorizes_total, df_fit['Total Paid Amount']))
df_fit['Number Of Paid Installments'] = list(map(categorizes_nr, df_fit['Number Of Paid Installments']))

df_fit.head()
```
